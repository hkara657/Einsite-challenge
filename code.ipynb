```python
from matplotlib import pyplot
import pandas as pa
import numpy as np
from datetime import datetime
from geopy.distance import great_circle
import math
```

```python
dataframe=pa.read_csv('train.csv')
print dataframe.dtypes

data=dataframe.values
print data.shape
nr,nc=data.shape

del dataframe  #clear memory
# def isfloat(value):
#   try:
#     float(value)
#     return True
#   except ValueError:
#     return False

# r,c=X_test.shape
# for i in np.arange(r):
#     for j in np.arange(c):
#         if isfloat(X_test[i,j]) and math.isnan(X_test[i,j]):
#             print i,j
```

```python
fmt='%Y-%m-%d %H:%M:%S'

pro_data=np.zeros((nr,11))

pro_data[:,10]=data[:,17]  #the final price to be predicted

b, c = np.unique(data[:,1], return_inverse=True)
pro_data[:,0]=c  #vendor id

b, c = np.unique(data[:,2], return_inverse=True)
pro_data[:,9]=c #whether new user

b, c = np.unique(data[:,11], return_inverse=True)
pro_data[:,7]=c #rate code

pro_data[:,3:6]=data[:,3:6]  # toll,tip,tax
pro_data[:,6]=data[:,8]  #passenger count
# pro_data[:,7]=data[:,11]  #rate_code
pro_data[:,8]=data[:,16]  #surcharge


for i in np.arange(nr):
    
    if i%10000==0:
        print i
    d1 = datetime.strptime(str(data[i,6]), fmt)
    d2 = datetime.strptime(str(data[i,7]), fmt)
    diff=d2-d1
    pro_data[i,1]=diff.total_seconds()/60.0   #total journey time
    
    loc1=(data[i,10],data[i,9])
    loc2=(data[i,14],data[i,13])
    pro_data[i,2]=great_circle(loc2,loc1).km  #total journey distance
    
```

```python
m1=np.nanmean(pro_data[:,1])
m2=np.nanmean(pro_data[:,2])
m4=np.nanmean(pro_data[:,4])
m5=np.nanmean(pro_data[:,5])
m8=np.nanmean(pro_data[:,8])

for i in np.arange(nr):
    
    if math.isnan(pro_data[i,1]):
        pro_data[i,1]=m1
        
    if math.isnan(pro_data[i,2]):
        pro_data[i,2]=m2
    
    if math.isnan(pro_data[i,4]):
        pro_data[i,4]=m4
        
    if math.isnan(pro_data[i,5]):
        pro_data[i,5]=m5
        
    if math.isnan(pro_data[i,8]):
        pro_data[i,8]=m8
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
X = pro_data[:,0:10]
Y = pro_data[:,10]
num_folds = 10
kfold = KFold(n_splits=10, random_state=7)
# model = Ridge()
# model=Lasso()
# model = DecisionTreeRegressor()
model = RandomForestRegressor(n_estimators=100, max_features=3)
# model=SVR()
scoring = 'neg_mean_absolute_error'
results = cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
print(results.mean())
```
